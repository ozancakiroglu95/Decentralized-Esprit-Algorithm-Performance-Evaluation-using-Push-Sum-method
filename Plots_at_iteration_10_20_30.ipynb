```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import bottleneck as bn

with open("RMSE_1000_samples_P_30.txt", "rb") as fp:   # Unpickling
    RMSE_1000_sample = pickle.load(fp)

mm = bn.move_mean(RMSE_1000_sample, window=40, min_count=1)

plt.plot(np.arange(2,1002), mm, 'r--')
plt.yscale("log")
plt.xlabel('Number of Samples N')
plt.ylabel('RMSE (degree)')
plt.show()
plt.plot(np.arange(2,1002), RMSE_1000_sample, 'r--')
plt.yscale("log")
plt.xlabel('Number of Samples N')
plt.ylabel('RMSE (degree)')
plt.show()
```
