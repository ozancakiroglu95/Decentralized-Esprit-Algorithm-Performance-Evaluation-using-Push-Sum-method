```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import bottleneck as bn
from matplotlib.pyplot import figure

figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

with open("Push_Sum_ESPRIT_1000_samples_P_10.txt", "rb") as fp:   # Unpickling
    push_sum_1000_sample_p_10 = pickle.load(fp)
    
with open("Push_Sum_ESPRIT_1000_samples_P_20.txt", "rb") as fp:   # Unpickling
    push_sum_1000_sample_p_20 = pickle.load(fp)
    
with open("Push_Sum_ESPRIT_1000_samples_P_30.txt", "rb") as fp:   # Unpickling
    push_sum_1000_sample_p_30 = pickle.load(fp)

with open("Average_Consensus_ESPRIT_1000_samples_P_10.txt", "rb") as fp:   # Unpickling
    ave_cons_1000_sample_p_10 = pickle.load(fp)
    
with open("Average_Consensus_ESPRIT_1000_samples_P_20.txt", "rb") as fp:   # Unpickling
    ave_cons_1000_sample_p_20 = pickle.load(fp)

with open("Average_Consensus_ESPRIT_1000_samples_P_30.txt", "rb") as fp:   # Unpickling
    ave_cons_1000_sample_p_30 = pickle.load(fp)

with open("Conventional_ESPRIT_1000_samples.txt", "rb") as fp:   # Unpickling
    conv_1000_sample_p_30 = pickle.load(fp)

plt.plot(np.arange(2,1002), push_sum_1000_sample_p_10, 'r-', label='Push-Sum Iteration:10')
plt.plot(np.arange(2,1002), push_sum_1000_sample_p_20, 'r--', label='Push-Sum Iteration:20')
plt.plot(np.arange(2,1002), push_sum_1000_sample_p_30, 'r-.', label='Push-Sum Iteration:30')
plt.plot(np.arange(2,1002), ave_cons_1000_sample_p_10, 'b-', label='Average Consensus Iteration:10')
plt.plot(np.arange(2,1002), ave_cons_1000_sample_p_20, 'b--', label='Average Consensus Iteration:20')
plt.plot(np.arange(2,1002), ave_cons_1000_sample_p_30, 'b-.', label='Average Consensus Iteration:30')
plt.plot(np.arange(2,1002), conv_1000_sample_p_30, 'g--', label='Conventional ESPRIT')
plt.legend()
plt.yscale("log")
plt.xlabel('Number of Samples N')
plt.ylabel('RMSE (degree)')
plt.show()
```
