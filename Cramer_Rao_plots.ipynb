```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import bottleneck as bn

with open("cramer_rao_1000.txt", "rb") as fp:   # Unpickling
    cramer_rao_1000 = pickle.load(fp)
    
with open("RMSE_1000_samples.txt", "rb") as fp:   # Unpickling
    RMSE_1000_sample = pickle.load(fp)

with open("RMSE_1000_samples_P_20.txt", "rb") as fp:   # Unpickling
    RMSE_1000_sample_P_20 = pickle.load(fp)

mm_20 = bn.move_mean(RMSE_1000_sample_P_20, window=10, min_count=1)
mm_10 = bn.move_mean(RMSE_1000_sample, window=10, min_count=1)

plt.plot(np.arange(2,1002), mm_20, 'r--', label='P=20')
plt.plot(np.arange(2,1002), mm_10, 'b--', label='P=10')
plt.plot(np.arange(2,1002), cramer_rao_1000, 'y--', label='Cramer_Rao')
plt.yscale("log")
plt.xlabel('Number of Samples N')
plt.ylabel('RMSE (degree)')
leg = plt.legend()
plt.show()
```
