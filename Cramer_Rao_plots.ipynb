```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import bottleneck as bn

with open("data/cramer_rao_data/Cramer_Rao_Bound_1000_Samples.txt", "rb") as fp:   # Unpickling
    cramer_rao_1000 = pickle.load(fp)

plt.plot(np.arange(2,1002), cramer_rao_1000, 'y--', label='Cramer_Rao')
plt.yscale("log")
plt.xlabel('Number of Samples N')
plt.ylabel('RMSE (degree)')
leg = plt.legend()
plt.show()
```
